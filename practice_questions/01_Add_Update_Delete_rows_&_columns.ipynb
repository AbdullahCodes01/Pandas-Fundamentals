```python
import pandas as pd
import numpy as np

emp_df = pd.DataFrame({
    "EmpID": [201,202,203,204,205,206,207,208,209,210],
    "Name": ["Aman","Riya","Kunal","Sneha","Vikram","Pooja","Rahul","Neha","Arjun","Simran"],
    "Department": ["IT","HR","IT","Finance","HR","IT","Finance","HR","IT","Finance"],
    "Age": [22,25,24,28,26,23,30,27,21,29],
    "Salary": [50000,45000,62000,70000,48000,58000,75000,52000,40000,68000],
    "Experience": [1,2,4,6,3,2,7,4,1,5]
})

emp_df
```

```python
# Add a new column Bonus with default value 5000
emp_df["bonus"] = 5000
```

```python
emp_df
```

```python
# Add a column Level

# Salary ≥ 65000 → "Senior"

# Salary 50000–64999 → "Mid"

# Salary < 50000 → "Junior"

emp_df["Level"] = "Junior"
emp_df.loc[emp_df["Salary"]>=65000,"Level"] = "Senior"
emp_df.loc[(emp_df["Salary"]>50000) & (emp_df["Salary"]<64999),"Level"] = "Mid"
emp_df
```

```python
new_row = pd.DataFrame([{
"EmpID": 211,
"Name": "Rohit",
"Department": "IT",
"Age": 26,
"Salary": 60000,
"Experience": 3,
"bonus": 5000,
"Level": "Mid"
}])

emp_df = pd.concat([emp_df,new_row],ignore_index=True)
emp_df
```

```python
# Increase Salary by 10% for employees from IT department

emp_df.loc[emp_df["Department"]== "IT","Salary"] *= 1.10
emp_df
```

```python
# Increase Bonus to 8000 for employees with Experience ≥ 5
emp_df.loc[emp_df["Experience"]>=5,"bonus"] = 8000
emp_df
```

```python
# Change Department of EmpID 205 to "Finance"
emp_df.loc[emp_df["EmpID"]==205,"Department"] = "Finance"
emp_df
```

```python
# Update Age of the youngest employee to 22
emp_df.loc[emp_df["Age"]== emp_df["Age"].min(),"Age"] = 22
emp_df

```

```python
# Delete the column Bonus
emp_df.drop("bonus",axis=1,inplace=True)
emp_df
```

```python
# Delete employees whose Salary < 45000
emp_df.drop(emp_df[emp_df["Salary"]<45000].index,inplace=True)
emp_df
```

```python
# Delete employees whose Experience < 2
emp_df.drop(emp_df[emp_df["Experience"]<2].index,inplace=True)
emp_df
```

```python
# Keep only employees from IT & Finance departments
emp_df = emp_df[emp_df["Department"].isin(["IT","Finance"])]
```

```python
emp_df
```

```python
print(emp_df["Department"].unique())
```

```python
# Add a column Status:

# Salary ≥ 60000 → "High Paid"

# Else → "Normal"

emp_df["Status"] = "Normal"
emp_df.loc[emp_df["Salary"]>=60000,"Status"] = "High Paid"
emp_df
```

```python
# Add column → df["col"] = value

# Conditional update → df.loc[condition, "col"] = value

# Delete rows → df.drop(df[condition].index, inplace=True)

# Delete column → df.drop("col", axis=1, inplace=True)
```

```python
# Reset index after deletions
emp_df.reset_index(drop=True,inplace=True)
emp_df
```

```python
# Sort DataFrame by Salary (descending)
emp_df.sort_values("Salary",ascending=False) # by default order is ascending order , for descending order ascending = false
emp_df
```
