4. series

What is a Series?

Series = 1-D labeled array
It is like one column of Excel


```python
import pandas as pd
import numpy as np
```

```python
s = pd.Series([85,88,76])
print(s)
```

```python
# (A) from List

l = pd.Series([10,20,30])
print(l)
print()

#( B) Custom index
c = pd.Series([30,40,50],index=['a','b','c'])
print(c)
print()

# (C) from Dictionary 
marks = {"ali": 40, "neha": 50, "zoya": 60}
m = pd.Series(marks)
print(m)
```

```python
m.index # to get index
m.values # to get values of series
m.dtype #datatype
```

```python
m['neha'] #Accessing elements by index label
```

```python
m[0] #accessing elements by index position
```

```python
# Multiple
m[0:2]
m[["ali","zoya"]]
```

```python
# Slicing & filtering
m[m<50]
m[m>50]
```

```python
#  Mathematical operations
# m + 5
# m*2
# m.mean()
# m.max
# m.min()

```

```python
# Missing values in Series
s = pd.Series([10,20,None,40])
s.isnull()
```

```python
s.notnull()
```

```python
s.fillna(0)
```

```python
s.dropna()
```

```python
# Useful Series methods
s.head()
s.tail()
s.unique()
s.value_counts()
s.sort_index()
s.sort_values()
```

```python
# practice question
# Create Series from dictionary of 5 students marks
students = {
    "Arman": 90,
    "farhan": 88,
    "sadiya": 79
}
s1 = pd.Series(students)
print(s1)
```

```python
# Print only marks > 80
print(s1[s1>80])

```

```python
# Add 10 marks to all
updated_s1 = s1 +10
print(updated_s1)
```

```python
updated_s1["rahul"] = np.nan
mean_val = updated_s1.mean()
updated_s1 = updated_s1.fillna(mean_val)
print(updated_s1)

```

```python
# Sort values descending
print(updated_s1.sort_values(ascending=False))
```
