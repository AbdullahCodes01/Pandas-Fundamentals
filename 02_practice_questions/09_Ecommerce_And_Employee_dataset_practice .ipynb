```python
# Find count of employees by Department + Gender combination.
df.value_counts(["Department","Gender"])
```

```python
# Convert value_counts output into DataFrame.
df["Department"].value_counts().reset_index(name="count")
```

```python
# Find which department company spends most salary on
# Combine groupby + count logic thinking
# df.groupby("Department")["Salary"].sum()
# pd.pivot_table(df,index="Department",values="Salary",aggfunc="sum").sort_index(ascending=False).head(1)
pd.pivot_table(df,index="Department",values="Salary",aggfunc="sum").sort_index(ascending=False).idxmax()
```

```python
# Find which city has most IT employees.
# df.groupby("City")["Name"].count().idxmax()
pd.pivot_table(df,index="City",values="Name",aggfunc="count").idxmax()
```

```python
# Find department distribution only for employees with experience ≥ 3.
df[df["Experience"]>=3]["Department"].value_counts(normalize=True)*100
```

```python
# using loc method
df.loc[df["Experience"] >= 3, "Department"].value_counts(normalize=True)*100
```

```python
# using crosstab method
pd.crosstab(df[df["Experience"] >= 3]["Department"], columns="Count",normalize=True)*100
```

```python
# Find % of male vs female employees.
df["Gender"].value_counts(normalize=True)*100
```

```python
pd.crosstab(df["Gender"],columns="Percentage",normalize=True)*100
```

```python
df2= pd.read_csv("/Users/armanshaikh/Desktop/pandas-fundamentals/practice_questions/orders_value_counts.csv")
```

```python
df2
```

```python
# Most used payment method.
df2["Payment_Method"].value_counts()
```

```python
# Order status distribution (Delivered / Cancelled / Returned).
df2["Order_Status"].value_counts()
```

```python
# Which city has most delivered orders.
df2[df2["Order_Status"]=="Delivered"]["City"].value_counts().idxmax()
```

```python
# Find % of each product category sold.
df2[df2["Order_Status"]=="Delivered"]["Category"].value_counts(normalize=True)*100
```

```python
# Find most common payment method for Delhi city only.
df2[df2["City"]=="Delhi"]["Payment_Method"].value_counts()
```

```python
# Create summary report:

# Find:
#  Most common Category
#  Most common city
#  Most common payment method
#  Most common order status

summary_report = {
    "category": df2["Category"].value_counts().idxmax(),
    "city": df2["City"].value_counts().idxmax(),
    "payment_method": df2["Payment_Method"].value_counts().idxmax(),
    "order_status": df2["Order_Status"].value_counts().idxmax()
}
summary_report
```

```python
report = pd.DataFrame(summary_report.items(),columns=["most common","result"])
```
