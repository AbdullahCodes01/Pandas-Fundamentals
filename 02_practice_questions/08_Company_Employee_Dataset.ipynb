```python
# Grouping
# Find:
# Average salary per department

df.groupby("department")["salary"].mean()
```

```python
pd.pivot_table(df,index="department",values="salary",aggfunc=["max","mean"])
```

```python
# Top 10 highest paid employees
pd.pivot_table(df,values="salary",index="name").sort_values(by="salary",ascending=False).head(10)
```

```python
# Employees with:
# Salary > company average
# Experience > 5 AND Salary > 8 LPA
df.loc[df["salary"]>df["salary"].mean()]
```

```python
# Index → department  
# Values → salary  
# Agg → mean

pd.pivot_table(df,values="salary",index="department",aggfunc="mean")
```
