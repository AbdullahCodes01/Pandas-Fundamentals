```python
import pandas as pd
import numpy as np
```

```python
# Employee basic info
employees = pd.DataFrame({
    "EmpID": [1,2,3,4,5],
    "Name": ["Aman","Riya","Kunal","Sneha","Vikram"],
    "DeptID": [101,102,101,103,102]
})

# Department details
departments = pd.DataFrame({
    "DeptID": [101,102,103],
    "Department": ["IT","HR","Finance"]
})

# Salary details
salary = pd.DataFrame({
    "EmpID": [1,2,3,5,6],
    "Salary": [50000,48000,62000,45000,70000]
})
```

```python
# Concatenate columns side-by-side
new_employees = pd.DataFrame({
    "EmpID": [6,7],
    "Name": ["Pooja","Rahul"],
    "DeptID": [101,103]
})

emp = pd.concat([employees,new_employees],axis=0)
emp.reset_index(drop=True)
```

```python
# Concatenate columns side-by-side
extra_info = pd.DataFrame({
    "Age": [22,25,24,28,26]
})
emp1 = pd.concat([emp.reset_index(drop=True),
           extra_info.reset_index(drop=True)], axis=1)

```

```python
# inner join emp1 and departments
# Show EmpID, Name, Department
emp2 = pd.merge(emp1,departments, on="DeptID",how="inner")
```

```python
# Left join employees with salary

# Keep all employees
# Missing salary should show NaN
emp3= pd.merge(emp2,salary,on="EmpID",how="left")
emp3
```

```python
# fter merging employees + departments + salary
# Create a final DataFrame with:
# Name
# Department
# Salary
result = emp3[["Name", "Department","Salary"]]
print(result)
```

```python
result.fillna(result["Salary"].median(),inplace=True)
result
```

```python
# Find departments where any employee salary > 60000
# result.loc[result["Salary"]>60000,"Department"].unique()
result.groupby("Department")["Salary"].max()>60000
```

```python
# Count number of employees in each department
result.groupby("Department")["Name"].count()
```
